### Drawing images and shapes using OpenCV

```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

def imshow(title = "", image = None, size = 10):
  h,w = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(aspect_ratio * size, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()

# download & unzip images
!wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
!unzip -qq images.zip
```

```python
# create a black image using numpy
image = np.zeros((512,512,3), np.uint8)

# Can we make this in black and white? Grayscale
gray_img = np.zeros((512,512), np.uint8)

imshow("black", image)
imshow("gray", gray_img)
```

##### Let's draw a line on black image

cv2.line(image, starting_cords, ending_cords, color, thickness)

```python
# Note this is the inplace operation, meaning it changes the input image
# Unlike many other OpenCV functions that return a new image leaving the input image unaffected.

cv2.line(image, (0,0), (511,511), (255, 127, 0), 5)

imshow("Black canvas with diagonal lines", image)
```

##### Drawing rectangles

cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)

```python
# create a black canvas
image = np.zeros((512,512,3), np.uint8)

# draw rectangle
cv2.rectangle(image, (110,130),(420,430), (255,127,0),5)

imshow("Rectangle on black canvas", image)
```

##### Drawing circle

cv2.circle(image, center, radius, color, fill)

```python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (350,350), 100, (15,150,50), -1)

imshow("Circle on blank canvas", image)
```

##### Drawing poygons

cv2.polylines(image, points, closed?, color, thickness)

```python
image = np.zeros((512,512,3), np.uint8)

# let's define four points
pts = np.array([[10,50], [400,50], [90,200], [50,500]], np.int32)

# reshape points in form required by ploygon
pts = pts.reshape(-1, 1, 2)

cv2.polylines(image, [pts], True, (255, 127,0), 5)

imshow("Ploygon", image)
```

##### Add text on image

cv2.putText(image, 'Text to display', bottom left starting point, Font, Font size, color, thickness)

Available fonts:

* FONT_HERSHEY_SIMPLEX = 0.
* FONT_HERSHEY_PLAIN = 1.
* FONT_HERSHEY_DUPLEX = 2.
* FONT_HERSHEY_COMPLEX = 3.
* FONT_HERSHEY_TRIPLEX = 4.
* FONT_HERSHEY_COMPLEX_SMALL = 5.
* FONT_HERSHEY_SCRIPT_SIMPLEX = 6.
* FONT_HERSHEY_SCRIPT_COMPLEX = 7.



```python
image = np.zeros((512,512,3), np.uint8)

text = "Hello Turab"

cv2.putText(image, text, (5,290), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 3, (255,127,0), 3)

imshow("Image with text", image)
```
