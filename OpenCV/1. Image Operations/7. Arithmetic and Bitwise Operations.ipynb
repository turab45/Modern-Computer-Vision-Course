### Arithmetic & Bitwise Operations

```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

def imshow(title = "", image = None, size = 10):
  h,w = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(aspect_ratio * size, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()

# download & unzip images
!wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
!unzip -qq images.zip
```

### Arithmetic Operations

```python
image = cv2.imread("/content/images/liberty.jpeg", 0)
imshow("Grayscale image", image)

# create a matrix of ones, then multiply it with scalar of 100
# This gives us a matrix with the same dimension as image and all values being 100

M = np.ones(image.shape, dtype="uint8") * 100
print(M)
```

#### Increase brightness

```python
added = cv2.add(image, M)

imshow("added image", added)

added2 = image + M

imshow("Added 2", added2)
```

### Decreasing brightness

```python
decreased = cv2.subtract(image, M)
imshow("Brightness decreased", decreased)

decreased2 = image - M
imshow("Brightness decreased2", decreased2)
```

### Bitwise Operations

```python
# Making an square
square = np.zeros((300,300), np.uint8)
cv2.rectangle(square, (50,50), (250,250), 255, -2)
imshow("Rectangle", square)

# Making an ellipse
ellipse = np.zeros((300,300), np.uint8)
cv2.ellipse(ellipse,(150,150), (150,150), 30, 0, 180, 255, -1)
imshow("Ellipse", ellipse)
```

```python
# Shows only where they intersect
And = cv2.bitwise_and(square, ellipse)
imshow("And", And)

# shows where either square or ellipse is
bitwiseOr = cv2.bitwise_or(square, ellipse)
imshow("Or", bitwiseOr)

# shows where either exist by itself
bitwiseXor = cv2.bitwise_xor(square, ellipse)
imshow("XOR", bitwiseXor)

# shows everything that isn't part of the square
bitwiseNot = cv2.bitwise_not(square)
imshow("NOT", bitwiseNot)

# 
```
