## Loading images

```python
img = cv2.imread("/content/images/flowers.jpeg")
```

## Displaying the image

```python
import matplotlib.pyplot as plt

# show the image with matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

Let's create a simple method to display our image in simpler and easier way

```python
def imshow(title="", image = None):
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
imshow("Flowers image", img)
```

## Saving image

```python
cv2.imwrite("output.jpeg", img)
```
