# Colour spaces - RGB and HSV

```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

def imshow(title = "", image = None, size = 10):
  h,w = image.shape[0], image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize=(aspect_ratio * size, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()

# download & unzip images
!wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
!unzip -qq images.zip
```

### 1. View the individual channels of RGB



```python
# load the image
img = cv2.imread("/content/images/castara.jpeg")

# split the image into colour channels
B, G, R = cv2.split(img)

imshow("Red channel only",image=R)
```

```python
# create a matrix of zeroes of same width and height as image
zeroes = np.zeros(img.shape[:2], dtype="uint8")

# display manipulated channels
imshow("Red channel manipulated", cv2.merge([zeroes, zeroes, R]))
imshow("Green channel manipulated", cv2.merge([zeroes, G, zeroes]))
imshow("Blue channel manipulated", cv2.merge([B, zeroes, zeroes]))
```

```python
# let's recreate the original image from three channels
img = cv2.imread("/content/images/castara.jpeg")

# split the image into colour channels
B, G, R = cv2.split(img)

# merge channels back to form the original image
merged = cv2.merge([B,G,R])

imshow("Merged image", merged)
```

### 2. Manipulate a colour space

```python
# let's amplify the blue color
merged = cv2.merge([B+100,G,R])

imshow("Amplified blue channel image", merged)
```

### 3. Introduce HSV colour space

```python
# reload the image
img = cv2.imread("/content/images/castara.jpeg")

# convert to HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

imshow("HSV image", hsv_img)
```

```python
# Let's view the representation of each HSV channel

imshow("Hue", hsv_img[:,:,0])
imshow("Saturation", hsv_img[:,:,1])
imshow("Value", hsv_img[:,:,2])
```
