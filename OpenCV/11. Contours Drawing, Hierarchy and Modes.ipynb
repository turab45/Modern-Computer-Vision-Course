# What are contours?
Contours are continues lines or curves that bound or cover the full boundary of an object.

````python
image = cv2.imread("/content/images/LP.jpg")
imshow("Image",image)
````

**Applying cv2.findContours()**


```
cv2.findContours(image, Retrieval Mode, Approximation method)
```
**Retrieval methods**
*  RETR_LIST
*  RETR_EXTERNAL
*  RETR_CCOMP
* RETR_TREE

**Approximation methods options**
* cv2.CHAIN_APPROX_NONE
* cv2.CHAIN_APPROX_SIMPLE



````python
image = cv2.imread("/content/images/LP.jpg", 0)

_,th2 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Finding contoues
contours, heirarchy = cv2.findContours(th2, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

# Draw all contours, note this alters the original image (inplace operation)
# use -
````
